```python
import numpy as np
from matplotlib import pyplot as plt
import cmath
```

## Question 1

### Fourier Series Representation of a Sine Wave

```python
def Fourier_Series_1(t_start, t_stop, n):
    t = np.arange(t_start, t_stop + 0.0001, 0.0001)
    fseries = 0
    for i in np.arange(1, n + 0.1, 1):
        fseries += np.sin(2 * np.pi * i * t)
    return fseries, t
```

```python
func1, t1 = Fourier_Series_1(0,1,1)

plt.figure(1)
plt.plot(t1, func1, 'r')
plt.title('Trigonometic Fourier Series Representation of a Sine Wave Function')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')
plt.show()
```

## Question 2

### Fourier Series Representation of a 50% Duty Cycle Square Wave

```python
def Fourier_Series_2(t_start, t_stop, n):
    t = np.arange(t_start, t_stop + 0.0001, 0.0001)
    fseries = 0
    for i in np.arange(1, n + 0.1, 1):
        fseries += 4/(np.pi*i) * np.sin((np.pi * i) / 2) * np.cos(2 * np.pi * i * t)
    return fseries, t
```

```python
func1, t1 = Fourier_Series_2(0,2,1)
func2, t2 = Fourier_Series_2(0,2,2)
func5, t5 = Fourier_Series_2(0,2,5)
func10, t10 = Fourier_Series_2(0,2,10)
G_Approx, t_Approx = Fourier_Series_2(0,2,100)

plt.figure(1)
plt.plot(t1, func1, 'r')
plt.title('Trigonometic Fourier Series Representation of 50% Duty Cycle Square Wave Function \n 1 Component')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(2)
plt.plot(t2, func2, 'b')
plt.title('Trigonometic Fourier Series Representation of 50% Duty Cycle Square Wave Function \n 2 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(3)
plt.plot(t5, func5, 'k')
plt.title('Trigonometic Fourier Series Representation of 50% Duty Cycle Square Wave Function \n 5 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(4)
plt.plot(t10, func10, 'g')
plt.title('Trigonometic Fourier Series Representation of 50% Duty Cycle Square Wave Function \n 10 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(5)
plt.plot(t_Approx, G_Approx, 'g')
plt.title('Trigonometic Fourier Series Representation of 50% Duty Cycle Square Wave Function \n 100 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.show()
```

## Question3

### Fourier Series Representation of a 10% Duty Cycle Square Wave

```python
def Fourier_Series_3(t_start, t_stop, n):
    t = np.arange(t_start, t_stop + 0.0001, 0.0001)
    fseries = 0
    Co = -0.8
    for i in np.arange(1, n + 0.1, 1):
        fseries += 4/(np.pi * i) * np.sin(0.1 * np.pi * i) * np.cos(2 * np.pi * i * t)
    result = Co + fseries
    return result, t
```

```python
func1, t1 = Fourier_Series_3(0,2,1)
func2, t2 = Fourier_Series_3(0,2,2)
func5, t5 = Fourier_Series_3(0,2,5)
func10, t10 = Fourier_Series_3(0,2,10)
G_Approx, t_Approx = Fourier_Series_3(0,2,100)

plt.figure(1)
plt.plot(t1, func1, 'r')
plt.title('Trigonometic Fourier Series Representation of 10% Duty Cycle Square Wave Function \n 1 Component')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(2)
plt.plot(t2, func2, 'b')
plt.title('Trigonometic Fourier Series Representation of 10% Duty Cycle Square Wave Function \n 2 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(3)
plt.plot(t5, func5, 'k')
plt.title('Trigonometic Fourier Series Representation of 10% Duty Cycle Square Wave Function \n 5 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(4)
plt.plot(t10, func10, 'g')
plt.title('Trigonometic Fourier Series Representation of 10% Duty Cycle Square Wave Function \n 10 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(5)
plt.plot(t_Approx, G_Approx, 'g')
plt.title('Trigonometic Fourier Series Representation of 10% Duty Cycle Square Wave Function \n 100 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.show()
```

## Question 4

### Fourier Series Representation of a Saw-Tooth Function

```python
def Fourier_Series_4(t_start, t_stop, n):
    t = np.arange(t_start, t_stop + 0.0001, 0.0001)
    fseries = 0
    Do = 0.5
    for i in np.arange(1, n + 0.1, 1):
        fseries += -(1/(np.pi * i)) * np.sin(2 * np.pi * i * t)
    result = Do + fseries
    return result, t
```

```python
func1, t1 = Fourier_Series_4(0,2,1)
func2, t2 = Fourier_Series_4(0,2,2)
func5, t5 = Fourier_Series_4(0,2,5)
func10, t10 = Fourier_Series_4(0,2,10)
G_Approx, t_Approx = Fourier_Series_4(0,2,100)

plt.figure(1)
plt.plot(t1, func1, 'r')
plt.title('Exponential Fourier Series Representation of Saw-Tooth Function \n 1 Component')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(2)
plt.plot(t2, func2, 'b')
plt.title('Exponential Fourier Series Representation of Saw-Tooth Function \n 2 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(3)
plt.plot(t5, func5, 'k')
plt.title('Exponential Fourier Series Representation of Saw-Tooth Function \n 5 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(4)
plt.plot(t10, func10, 'g')
plt.title('Exponential Fourier Series Representation of Saw-Tooth Function \n 10 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.figure(5)
plt.plot(t_Approx, G_Approx, 'g')
plt.title('Exponential Fourier Series Representation of Saw-Tooth Function \n 100 Components')
plt.xlabel('Frequency [rad/s]')
plt.ylabel('Amplitude')

plt.show()
```
